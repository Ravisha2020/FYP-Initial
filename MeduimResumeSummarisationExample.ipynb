```python
import re
import subprocess

PHONE_REG = re.compile(r'[\+\(]?[1-9][0-9 .\-\(\)]{8,}[0-9]')


def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)


def extract_phone_number(resume_text):
    phone = re.findall(PHONE_REG, resume_text)

    if phone:
        number = ''.join(phone[0])

        if resume_text.find(number) >= 0 and len(number) <16:
            return number
    return None


if __name__ == '__main__':
    text = extract_text_from_pdf('resume-sample.pdf')
    phone_number = extract_phone_number(text)

    print(phone_number)
```

```python
import docx2txt
import nltk

nltk.download('stopwords')
nltk.download('punkt')

# you may read the database from a csv file or some other database
SKILLS_DB = [
    'machine learning', 'data science', 'python', 'word', 'excel', 'English', 'Active Directory',
    'Azure', 'anti-virus', 'Backup Exec', 'backup', 'Bash', 'batch', 'Cacti', 'Cisco ASA', 'databases',
    'DHCP', 'DNS', 'documentation', 'DataDomain', 'EMC', 'Enterprise Vault',
    'ePO', 'file servers', 'firewall', 'GPO', 'HTML', 'IIS', 'ISA', 'LDAP', 'Linux', 'McAfee', 'Exchange',
    'Microsoft Office',  'Microsoft Windows', 'security', 'policies', 'PowerShell', 'programming', 'proxy server',
    'servers', 'scripts', 'SolarWinds', 'SQL', 'StorSimple', 'troubleshooting', 'TMG', 'Ubuntu',
    'Visual Basic Script', 'VBS', 'Veritas Netbackup', 'VPN', 'VRanger', 'Veeam', 'VMWare',
    'VDI', 'virtual manchine', 'NMap', 'ZenMap'
]


def extract_text_from_docx(docx_path):
    txt = docx2txt.process(docx_path)
    if txt:
        return txt.replace('\t', ' ')
    return None


def extract_skills(input_text):
    stop_words = set(nltk.corpus.stopwords.words('english'))
    word_tokens = nltk.tokenize.word_tokenize(input_text)

    # remove the stop words
    filtered_tokens = [w for w in word_tokens if w not in stop_words]

    # remove the punctuation
    filtered_tokens = [w for w in word_tokens if w.isalpha()]

    # generate bigrams and trigrams (such as artificial intelligence)
    bigrams_trigrams = list(map(' '.join, nltk.everygrams(filtered_tokens, 2, 3)))

    # we create a set to keep the results in.
    found_skills = set()

    # we search for each token in our skills database
    for token in filtered_tokens:
        if token.lower() in SKILLS_DB:
            found_skills.add(token)

    # we search for each bigram and trigram in our skills database
    for ngram in bigrams_trigrams:
        if ngram.lower() in SKILLS_DB:
            found_skills.add(ngram)

    return found_skills


if __name__ == '__main__':
    text = extract_text_from_pdf('resume.pdf')
    skills = extract_skills(text)

    print(skills)
```

```python
import docx2txt
import nltk

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')


RESERVED_WORDS = [
    'school',
    'college',
    'univers',
    'academy',
    'faculty',
    'institute',
    'faculdades',
    'Schola',
    'schule',
    'lise',
    'lyceum',
    'lycee',
    'polytechnic',
    'kolej',
    'ünivers',
    'okul',
    'Univeristy'
]

def extract_education(input_text):
    organizations = []

    # first get all the organization names using nltk
    for sent in nltk.sent_tokenize(input_text):
        for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent))):
            if hasattr(chunk, 'label') and chunk.label() == 'ORGANIZATION':
                organizations.append(' '.join(c[0] for c in chunk.leaves()))

    # we search for each bigram and trigram for reserved words
    # (college, university etc...)
    education = set()
    for org in organizations:
        for word in RESERVED_WORDS:
            if org.lower().find(word) >= 0:
                education.add(org)

    return education


if __name__ == '__main__':
    text = extract_text_from_pdf('resume.pdf')
    education_information = extract_education(text)

    print(education_information)
```

```python
import re

def summarize_resume_section(resume_text, section_of_interest):
    # Use regular expression to find the section of interest
    pattern = re.compile(f"{section_of_interest}\s*([\s\S]*?)(?=\n\n|\n\w|$)", re.IGNORECASE)
    match = re.search(pattern, resume_text)

    if not match:
        return "Section not found in the resume."

    relevant_section = match.group(1).strip()

    return relevant_section

# Example usage
resume_text = extract_text_from_pdf('./functionalsample.pdf')

section_of_interest = "Childcare Experience"
result = summarize_resume_section(resume_text, section_of_interest)
print(result)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multioutput import MultiOutputClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import MultiLabelBinarizer

# Load labeled data from CSV file
df = pd.read_csv('labeled_data.csv')

# Split the dataset
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Separate features and labels
X_train, y_train = train_data['Sentence'], [categories.split(',') for categories in train_data['Categories']]
X_test, y_test = test_data['Sentence'], [categories.split(',') for categories in test_data['Categories']]

# Convert labels to binary format for multi-label classification
mlb = MultiLabelBinarizer()
y_train_binary = mlb.fit_transform(y_train)
y_test_binary = mlb.transform(y_test)

# Create a pipeline with TF-IDF vectorizer and multi-label SVM classifier
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('svm', MultiOutputClassifier(SVC(kernel='linear')))
])

# Train the model
model.fit(X_train, y_train_binary)

# Test the model
predictions = model.predict(X_test)

# Evaluate the model
# accuracy = accuracy_score(y_test_binary, predictions)
# print(f"Model Accuracy: {accuracy}")

# Display classification report
# print("Classification Report:")
# print(classification_report(y_test_binary, predictions, target_names=mlb.classes_))
```

```python
import torch
from transformers import AutoTokenizer, AutoModel
from textblob import TextBlob

# Load BERT tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

# Calculate the cosine similarity between two texts
def cosine_similarity(text1, text2):
    # Tokenize the texts and convert to tensors
    tokens1 = tokenizer(text1, return_tensors="pt")
    tokens2 = tokenizer(text2, return_tensors="pt")

    # Get the embeddings for the tokens
    embeddings1 = model(**tokens1).last_hidden_state.mean(dim=1)
    embeddings2 = model(**tokens2).last_hidden_state.mean(dim=1)

    # Calculate the cosine similarity
    dot_product = torch.matmul(embeddings1, embeddings2.transpose(1, 0))
    norm1 = torch.norm(embeddings1, dim=1)
    norm2 = torch.norm(embeddings2, dim=1)
    cosine_similarity = dot_product / (norm1 * norm2)

    return cosine_similarity.item()

# Test the function
text1 = "Languages"
text2 = "Programming"
text3 = "The dog barked at the moon"

textBlobText1 = TextBlob(text1)
textBlobText2 = TextBlob(text2)

spellCheckedText1 = str(textBlobText1.correct())
spellCheckedText2 = str(textBlobText2.correct())

similarity1 = cosine_similarity(spellCheckedText1, spellCheckedText2)
similarity2 = cosine_similarity(text1, text3)

print(f"Similarity between {spellCheckedText1} and {spellCheckedText2}: {similarity1}")
print(f"Similarity between text1 and text3: {similarity2:.2f}")
```
